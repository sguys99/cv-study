### 1. VIT 개요

이미지 → 패치 분할 → Linear Embedding → Transformer Encoder → Classification Head

- 이미지를 고정 크기 패치(16x16 또는 14x14)로 분할
- 각 패치를 토큰처럼 취급
- NLP Transformer 아키텍처 그대로 적용
